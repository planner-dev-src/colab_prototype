```python
# ============================================================
# БЛОК 3: ЗАГРУЗКА КОНФИГУРАЦИИ
# ============================================================
import json
import os

config_path = os.path.join(PROJECT_ROOT, 'config', 'metrics_registry.json')

try:
    with open(config_path, 'r', encoding='utf-8') as f:
        metrics_config = json.load(f)
    print(f"✅ Конфигурация загружена: {len(metrics_config)} метрик")
    print(f"   Путь: {config_path}")
except FileNotFoundError:
    print(f"⚠️ Файл конфигурации не найден: {config_path}")
    metrics_config = None
```

```python
# ============================================================
# БЛОК 4: ЗАПУСК АГЕНТА РЫНКА (ШАГ 1)
# ============================================================
print("=" * 60)
print("📊 ШАГ 1: АГЕНТ РЫНКА")
print("=" * 60)

market_agent = MarketAgent()
market_result = market_agent.run()

# Извлекаем данные из AgentResult
if market_result.success:
    data = market_result.data
    platforms_count = data.get('platforms_count', 0)
    total_courses = data.get('total_courses', 0)
    platforms = data.get('platforms', [])
    
    print(f"  Платформ обработано: {platforms_count}")
    print(f"  Курсов собрано: {total_courses}")
    print(f"  Платформы: {', '.join(platforms)}")
    print("✅ Агент рынка завершил работу")
else:
    print(f"❌ Ошибка работы агента: {market_result.error}")
```

```python
# ============================================================
# БЛОК 5: ЗАПУСК АГЕНТА АНАЛИЗА РЫНКА (ШАГ 2)
# ============================================================
print("=" * 60)
print("📊 ШАГ 2: АНАЛИЗ РЫНКА")
print("=" * 60)

analysis_result = run_market_analysis()

# Извлекаем данные из словаря
if analysis_result:
    summary = analysis_result.get("summary", {})
    platforms_total = summary.get('platforms_total', 0)
    trend_signals_total = summary.get('trend_signals_total', 0)
    competitive_gaps_total = summary.get('competitive_gaps_total', 0)
    
    print(f"  Платформ: {platforms_total}")
    print(f"  Трендов: {trend_signals_total}")
    print(f"  Gap-зон: {competitive_gaps_total}")
    
    # Показываем детали
    platforms = analysis_result.get('platforms', [])
    if platforms:
        print(f"\n  📋 Платформы:")
        for p in platforms:
            print(f"    • {p.get('name')}: {p.get('courses', 0)} курсов")
    
    trends = analysis_result.get('trends', [])
    if trends:
        print(f"\n  📈 Тренды:")
        for t in trends:
            print(f"    • {t.get('name')}: {t.get('strength', 'unknown')}")
    
    gaps = analysis_result.get('gaps', [])
    if gaps:
        print(f"\n  🎯 GAP-зоны:")
        for g in gaps:
            print(f"    • {g.get('topic')}: opportunity {g.get('opportunity', 0):.2f}")
    
    print("✅ Анализ рынка завершён")
else:
    print("❌ Ошибка: анализ рынка не выполнен")
```

```python
# ============================================================
# БЛОК 6: ЗАПУСК АГЕНТА МЕТРИК (ШАГ 3)
# ============================================================
print("=" * 60)
print("📈 ШАГ 3: АГЕНТ МЕТРИК")
print("=" * 60)

metrics_agent = MetricsAgent()
metrics_result = metrics_agent.run()

# Извлекаем данные из AgentResult
if metrics_result.success:
    data = metrics_result.data
    metrics = data.get('metrics', [])
    total = data.get('total', 0)
    critical = data.get('critical', 0)
    warning = data.get('warning', 0)
    summary = data.get('summary', {})
    
    print(f"  Всего метрик: {total}")
    print(f"  Критических: {critical}")
    print(f"  Предупреждений: {warning}")
    
    # Показываем статусы
    status_distribution = summary.get('by_status', {})
    if status_distribution:
        print(f"\n  📊 Распределение по статусам:")
        for status, count in status_distribution.items():
            print(f"    • {status}: {count}")
    
    # Показываем критические метрики
    if critical > 0:
        print(f"\n  🔴 Критические метрики:")
        critical_metrics = metrics_agent.get_critical_metrics()
        for m in critical_metrics:
            print(f"    • {m.name}: {m.value} ({m.interpretation})")
    
    print("✅ Агент метрик завершил работу")
else:
    print(f"❌ Ошибка работы агента: {metrics_result.error}")
```

```python
# ============================================================
# БЛОК 7: ФОРМИРОВАНИЕ ОТЧЁТА (ШАГ 4)
# ============================================================
print("=" * 60)
print("📄 ШАГ 4: ФОРМИРОВАНИЕ ОТЧЁТА")
print("=" * 60)

report_agent = ReportAgent()

# Подготавливаем данные для отчёта
report_data = {
    "market_analysis": analysis_result,
    "metrics": metrics_agent
}

report_result = report_agent.generate_report(report_data)

if report_result:
    recommendations = report_result.get('recommendations', {})
    strategic_recs = recommendations.get('strategic', [])
    
    print(f"✅ Отчёт сформирован!")
    print(f"  Рекомендаций: {len(strategic_recs)}")
    
    # Показываем статус
    status = report_result.get('status', 'unknown')
    print(f"  Статус: {status}")
else:
    print("❌ Ошибка формирования отчёта")
```

```python
# ============================================================
# БЛОК 8: ВЫВОД ИТОГОВОГО ОТЧЁТА
# ============================================================
print("\n" + "=" * 60)
print("📋 ИТОГОВЫЙ ОТЧЁТ")
print("=" * 60)

if report_result:
    print("\n📌 РЕЗЮМЕ:")
    print(report_result.get("executive_summary", "Нет данных"))
    
    print("\n📌 РЕКОМЕНДАЦИИ:")
    recommendations = report_result.get("recommendations", {})
    
    for level, recs in recommendations.items():
        if recs:
            print(f"\n  {level.upper()}:")
            for i, rec in enumerate(recs[:5], 1):
                print(f"    {i}. {rec}")
    
    # Дополнительная информация
    if 'market_analysis' in report_result:
        print("\n📊 АНАЛИЗ РЫНКА:")
        market_analysis = report_result['market_analysis']
        summary = market_analysis.get('summary', {})
        print(f"  Платформ: {summary.get('platforms_total', 0)}")
        print(f"  Трендов: {summary.get('trend_signals_total', 0)}")
        print(f"  GAP-зон: {summary.get('competitive_gaps_total', 0)}")
    
    if 'metrics' in report_result:
        metrics_info = report_result['metrics']
        print(f"\n📈 МЕТРИКИ:")
        print(f"  Всего: {metrics_info.get('total', 0)}")
        print(f"  Критических: {metrics_info.get('critical', 0)}")
else:
    print("❌ Отчёт не сформирован")

print("\n" + "=" * 60)
print("✅ ПРОТОТИП ЗАВЕРШЁН")
print("=" * 60)
```

```python
# ============================================================
# БЛОК 9: СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
# ============================================================
print("\n" + "=" * 60)
print("💾 СОХРАНЕНИЕ РЕЗУЛЬТАТОВ")
print("=" * 60)

import json
import os
from datetime import datetime

# Определяем папку для сохранения
reports_dir = os.path.join(PROJECT_ROOT, 'reports')
os.makedirs(reports_dir, exist_ok=True)
print(f"📁 Папка для отчётов: {reports_dir}")

# Функция для сериализации объектов в JSON
def to_serializable(obj):
    if hasattr(obj, 'to_dict'):
        return obj.to_dict()
    elif hasattr(obj, '__dict__'):
        return {k: to_serializable(v) for k, v in obj.__dict__.items() if not k.startswith('_')}
    elif isinstance(obj, dict):
        return {k: to_serializable(v) for k, v in obj.items()}
    elif isinstance(obj, (list, tuple)):
        return [to_serializable(item) for item in obj]
    elif isinstance(obj, (str, int, float, bool, type(None))):
        return obj
    else:
        return str(obj)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# ====== СОХРАНЯЕМ JSON ======
try:
    report_data = {
        "timestamp": timestamp,
        "execution_time": datetime.now().isoformat(),
        "report": to_serializable(report_result) if report_result else {},
        "market_analysis": to_serializable(analysis_result) if analysis_result else {},
        "metrics_summary": to_serializable(metrics_agent.get_summary()) if metrics_agent else {},
        "environment": {
            "python_version": sys.version,
            "platform": sys.platform,
            "cwd": os.getcwd()
        }
    }
    
    json_path = os.path.join(reports_dir, f"report_{timestamp}.json")
    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(report_data, f, ensure_ascii=False, indent=2)
    
    print(f"✅ JSON-отчёт сохранён: {json_path}")
    
except Exception as e:
    print(f"❌ Ошибка сохранения JSON: {e}")

# ====== СОХРАНЯЕМ ТЕКСТОВУЮ ВЕРСИЮ ======
try:
    txt_path = os.path.join(reports_dir, f"report_{timestamp}.txt")
    with open(txt_path, 'w', encoding='utf-8') as f:
        f.write("=" * 60 + "\n")
        f.write("ОТЧЁТ ПРОТОТИПА НЕЙРО-СОТРУДНИКА\n")
        f.write("=" * 60 + "\n\n")
        
        f.write(f"Время выполнения: {timestamp}\n\n")
        
        f.write("📌 РЕЗЮМЕ:\n")
        if report_result:
            f.write(report_result.get("executive_summary", "Нет данных") + "\n\n")
        else:
            f.write("Отчёт не сформирован\n\n")
        
        f.write("📌 РЕКОМЕНДАЦИИ:\n")
        if report_result:
            recommendations = report_result.get("recommendations", {})
            if recommendations:
                for level, recs in recommendations.items():
                    if recs:
                        f.write(f"\n{level.upper()}:\n")
                        for rec in recs[:5]:
                            f.write(f"  • {rec}\n")
            else:
                f.write("  Нет рекомендаций\n")
        else:
            f.write("  Нет рекомендаций\n")
        
        if metrics_agent:
            f.write("\n" + "=" * 60 + "\n")
            f.write("📊 МЕТРИКИ:\n")
            summary = metrics_agent.get_summary()
            f.write(f"  Всего метрик: {summary.get('total', 0)}\n")
            f.write(f"  Критических: {summary.get('critical_count', 0)}\n")
            f.write(f"  Предупреждений: {summary.get('warning_count', 0)}\n")
        
        if analysis_result:
            f.write("\n" + "=" * 60 + "\n")
            f.write("📊 АНАЛИЗ РЫНКА:\n")
            summary = analysis_result.get('summary', {})
            f.write(f"  Платформ: {summary.get('platforms_total', 0)}\n")
            f.write(f"  Трендов: {summary.get('trend_signals_total', 0)}\n")
            f.write(f"  GAP-зон: {summary.get('competitive_gaps_total', 0)}\n")
        
        f.write("\n" + "=" * 60 + "\n")
        f.write(f"✅ Отчёт сохранён: {timestamp}\n")
    
    print(f"✅ Текстовая версия сохранена: {txt_path}")
    
except Exception as e:
    print(f"⚠️ Ошибка сохранения текстовой версии: {e}")

# ====== ОПЦИОНАЛЬНО: КОПИРУЕМ НА GOOGLE DISK ======
try:
    import google.colab
    IN_COLAB = True
except ImportError:
    IN_COLAB = False

if IN_COLAB:
    try:
        from google.colab import drive
        drive.mount('/content/drive')
        
        # Копируем отчёты на диск
        !mkdir -p "/content/drive/MyDrive/colab_reports"
        !cp -r "{reports_dir}/"* "/content/drive/MyDrive/colab_reports/"
        
        print(f"✅ Отчёты скопированы на Google Disk: /content/drive/MyDrive/colab_reports/")
    except Exception as e:
        print(f"⚠️ Не удалось скопировать на Google Disk: {e}")

print("\n" + "=" * 60)
print("✅ СОХРАНЕНИЕ ЗАВЕРШЕНО")
print("=" * 60)
```
